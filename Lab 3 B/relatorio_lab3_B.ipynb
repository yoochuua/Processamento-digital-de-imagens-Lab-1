```python
#Constantes:
PATH = "/content/drive/MyDrive/PDI-2026/Lab3/"
IMG_ORIGINAL = cv2.imread(f'{PATH}foto1.jpg')
IMG_AVATAR = cv2.imread(f'{PATH}foto_personagens.png')
SOBEL_X = 'x'
SOBEL_Y =   'y'
KERNEL = [1, 3, 5]
LIMITES = [(50, 150), (100, 200), (30, 90)]
if IMG_ORIGINAL is None:
    print("Erro: imagem não carregada!")
```

Abaixo temos as funções dos filtros utilizados nessa atividade:

```python
def filtro_laplaciano(img):
  escala_cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  laplaciano = cv2.Laplacian(escala_cinza, cv2.CV_64F)
  laplaciano = cv2.convertScaleAbs(laplaciano)
  return laplaciano

```

```python
def filtro_sobel(img, tipo, kernel):
  cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
  if tipo == 'x':
    sobel = cv2.Sobel(cinza, cv2.CV_64F, 1, 0, ksize=kernel)
  else:
    sobel = cv2.Sobel(cinza, cv2.CV_64F, 0, 1, ksize=kernel)
  sobel = cv2.convertScaleAbs(sobel)
  return sobel
```

```python
def filtro_canny(img, inferior, superior, kernel):
  cinza = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY,)
  canny = cv2.Canny(cinza, inferior, superior, aperture_size=kernel)
  return canny
```

```python
def filtro_equacao_gonzalez(img, c):
  return img + (c * filtro_laplaciano(img))
```

```python
def fitro_highboost(img, n_realce):
  blur = cv2.GaussianBlur(img, (5,5), 0)
  mascara = img - blur
  highboost = img + n_realce * mascara
  highboost = cv2.convertScaleAbs(highboost)
  return highboost
```

### Item 1a:

```python
def item_1a(img, nome_img):

  cv2.imwrite(f'{nome_img}_laplaciano.jpg', filtro_laplaciano(img))

  for k in KERNEL:
    cv2.imwrite(f'{nome_img}_sobel_x_{k}x{k}.jpg', filtro_sobel(img, SOBEL_X, k))
    cv2.imwrite(f'{nome_img}_sobel_y_{k}x{k}.jpg', filtro_sobel(img, SOBEL_Y, k))

  i=0
  for x, y in LIMITES:
    cv2.imwrite(f'{nome_img}_canny_{x}x{y}.jpg', filtro_canny(img, x, y, KERNEL[i]))
    i = i + 1
```

```python
item_1a(IMG_ORIGINAL, 'img_original')
item_1a(IMG_AVATAR, 'img_avatar')
```

### Item 2a:

```python
def item_2a(img, nome_img):
  cv2.imwrite(f'{nome_img}_gonzalez.jpg', filtro_equacao_gonzalez(img, -1))

  img = cv2.imread(f'{PATH}{nome_img}_laplaciano.jpg')
  cv2.imwrite(f'{nome_img}_laplaciano_gonzalez.jpg', filtro_equacao_gonzalez(img, -1))


  for k in KERNEL:
    img = cv2.imread(f'{PATH}{nome_img}_sobel_x_{k}x{k}.jpg')
    cv2.imwrite(f'{nome_img}_sobel_x_{k}x{k}_gonzalez.jpg', filtro_equacao_gonzalez(img, -1))
    img = cv2.imread(f'{PATH}{nome_img}_sobel_y_{k}x{k}.jpg')
    cv2.imwrite(f'{nome_img}_sobel_y_{k}x{k}_gonzalez.jpg', filtro_equacao_gonzalez(img, -1))

  for x, y in LIMITES:
    img = cv2.imread(f'{PATH}{nome_img}_canny_{x}x{y}.jpg')
    cv2.imwrite(f'{nome_img}_canny_{x}x{y}_gonzalez.jpg', filtro_equacao_gonzalez(img, -1))

```
