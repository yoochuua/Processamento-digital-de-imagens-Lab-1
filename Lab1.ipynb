## Imagens invertidas de rgb para tons de cinza

### Histograma e Equalização para imagem do Van Gogh

```python
src = cv.cvtColor(image_rgb, cv.COLOR_BGR2GRAY)
cv2_imshow(src)
```

Mostra o histograma com a distribuição do contraste da imagem

```python
hist = cv.calcHist([src], [0], None, [256], [0, 256])
plt.figure(figsize=(8, 4))
plt.bar(range(256), hist.flatten(), color='blue', edgecolor = 'gray', width=1.0)
plt.title('Histograma Van Gogh (H)')
plt.xlabel('Intensidade (0-255)')
plt.ylabel('Quantidade de Pixels')
plt.xlim([-1, 256])
#plt.ylim([0, 140000])
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

### Equalização da imagem
  Este método melhora o contraste em uma imagem, a fim de esticar o intervalo de intensidade.

  A partir do histogrma acima, é possível perceber que os pixels parecem agrupados em torno do meio da gama disponível de intensidades.
  
  O que a **Equalização Histograma** faz é esticar esse intervalo. A imagem parece mais nítida porque a equalização mapeia os tons originais para que eles ocupem de forma mais uniforme todo o espectro de intensidade disponível (de 0 a 255). Isso amplia a diferença tonal entre objetos vizinhos que antes tinham cores muito similares, destacando detalhes e texturas que estavam 'escondidos' no baixo contraste.

```python
dst = cv.equalizeHist(src)
cv2_imshow(dst)
```

```python
hist = cv.calcHist([dst], [0], None, [256], [0, 256])
plt.figure(figsize=(8, 4))
plt.bar(range(256), hist.flatten(), color='blue', edgecolor = 'gray', width=1.0)
plt.title('Histograma Van Gogh (H)')
plt.xlabel('Intensidade (0-255)')
plt.ylabel('Quantidade de Pixels')
plt.xlim([-1, 256])
#plt.ylim([0, 140000])
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.tight_layout()
plt.show()
```

# Analisando o histograma em imagens coloridas

```python
image = io.imread(van_gogh)
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
cv2_imshow(image_rgb)
```

```python
bgr_planes = cv.split(image_rgb)

histSize = 256

histRange = (0, 256) # the upper boundary is exclusive


accumulate = False


b_hist = cv.calcHist(bgr_planes, [0], None, [histSize], histRange, accumulate=accumulate)
g_hist = cv.calcHist(bgr_planes, [1], None, [histSize], histRange, accumulate=accumulate)
r_hist = cv.calcHist(bgr_planes, [2], None, [histSize], histRange, accumulate=accumulate)


hist_w = 512
hist_h = 400
bin_w = int(round( hist_w/histSize ))

histImage = np.zeros((hist_h, hist_w, 3), dtype=np.uint8)


cv.normalize(b_hist, b_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)
cv.normalize(g_hist, g_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)
cv.normalize(r_hist, r_hist, alpha=0, beta=hist_h, norm_type=cv.NORM_MINMAX)



for i in range(1, histSize):
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(b_hist[i-1][0]) ),
            ( bin_w*(i), hist_h - int(b_hist[i][0]) ),
            ( 255, 0, 0), thickness=2)
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(g_hist[i-1][0]) ),
            ( bin_w*(i), hist_h - int(g_hist[i][0]) ),
            ( 0, 255, 0), thickness=2)
    cv.line(histImage, ( bin_w*(i-1), hist_h - int(r_hist[i-1][0]) ),
            ( bin_w*(i), hist_h - int(r_hist[i][0]) ),
            ( 0, 0, 255), thickness=2)


cv2_imshow(histImage)

```
